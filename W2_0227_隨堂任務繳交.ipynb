```python
food_calories = {
    "grains": {
        "Rice": 200,
        "Noodles": 220
    },
    "proteins": {
        "Egg": 70,
        "Tofu": 80
    }
}
```

```python
food_calories['proteins'].keys()
```

```python
import pandas as pd

df = pd.DataFrame(list(food_calories.keys()), columns=['Item'])
```

```python
df
```

```python
# 轉換為長格式
data_all = []
for category, items in food_calories.items():
    for food, calories in items.items():
        data_all.append([category, food, calories])

# 建立 DataFrame
df_all = pd.DataFrame(data_all, columns=['Category', 'Food', 'Calories'])
```
